```python


import matplotlib.ticker as ticker

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Total Household Income'])
plt.xlabel('Index')
plt.ylabel('Total Household Income')
plt.title('Line Graph of Total Household Income')

# Format y-axis labels for better readability (e.g., thousands)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}K'.format(x/1000)))

plt.show()
```

```python
# prompt: draw a scatter plot for Region and total household income

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Region', y='Total Household Income', data=df)
plt.xlabel('Region')
plt.ylabel('Total Household Income')
plt.title('Scatter Plot of Region vs. Total Household Income')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# prompt: plot total household income vs total food expenditure

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Household Income', y='Total Food Expenditure', data=df)
plt.xlabel('Total Household Income')
plt.ylabel('Total Food Expenditure')
plt.title('Scatter Plot of Total Household Income vs. Total Food Expenditure')
# Format y-axis labels for better readability (e.g., thousands)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}K'.format(x/1000)))
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}K'.format(x/1000)))
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
# prompt: generate a line chart with Total Number of Family members on x axis and the average of Total Household Income for that age on y axis

# Calculate the average Total Household Income for each number of family members
average_income_by_family_size = df.groupby('Total Number of Family members')['Total Household Income'].mean()

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(average_income_by_family_size.index, average_income_by_family_size.values)
plt.xlabel('Total Number of Family Members')
plt.ylabel('Average Total Household Income')
plt.title('Line Chart of Family Size vs. Average Total Household Income')

# Format y-axis labels for better readability (e.g., thousands)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}K'.format(x/1000)))

plt.show()
```

```python
# prompt: generate a line chart with Total number of family members employed on x axis and the average of Total Household Income for that age on y axis

# Calculate the average Total Household Income for each number of family members employed
average_income_by_employed = df.groupby('Total number of family members employed')['Total Household Income'].mean()

# Create the line chart
plt.figure(figsize=(10, 6))
plt.plot(average_income_by_employed.index, average_income_by_employed.values)
plt.xlabel('Total Number of Family Members Employed')
plt.ylabel('Average Total Household Income')
plt.title('Line Chart of Family Members Employed vs. Average Total Household Income')

# Format y-axis labels for better readability (e.g., thousands)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}K'.format(x/1000)))

plt.show()
```

```python
# prompt: draw a area chart with Type of Household on x axis, the average Total Household Income on y axis and have different colors of area as per Household Head Sex

# Calculate the average Total Household Income for each Type of Household and Household Head Sex
average_income = df.groupby(['Type of Household', 'Household Head Sex'])['Total Household Income'].mean().reset_index() # Use reset_index() here instead of unreset_index()

# Pivot the data to have Household Head Sex as columns
pivot_df = average_income.pivot(index='Type of Household', columns='Household Head Sex', values='Total Household Income')

# Create the area chart
plt.figure(figsize=(12, 6))
pivot_df.plot.area(stacked=False)
plt.xlabel('Type of Household')
plt.ylabel('Average Total Household Income')
plt.title('Area Chart of Type of Household vs. Average Total Household Income by Household Head Sex')

# Format y-axis labels for better readability (e.g., thousands)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}K'.format(x/1000)))

plt.legend(title='Household Head Sex')
plt.show()
```

```python
# prompt: make a waterfall chart showing green for increase and red for decrease, columns for increase:Total Household Income,Total Income from Entrepreneurial Acitivites. columns for decrease: all columns having 'Expenditure' as last word.Do it for the first row of data only.

import waterfall_chart

# Extract data for the first row
row_data = df.iloc[0]

# Define increase and decrease columns
increase_cols = ['Total Household Income', 'Total Income from Entrepreneurial Acitivites']
decrease_cols = [col for col in df.columns if col.endswith('Expenditure')]

# Extract values for increase and decrease
increase_values = [row_data[col] for col in increase_cols]
decrease_values = [-row_data[col] for col in decrease_cols]  # Negative for decrease

# Combine data and labels
values = increase_values + decrease_values
labels = increase_cols + decrease_cols

# Create waterfall chart
plt.figure(figsize=(12, 6))
waterfall_chart.plot(labels, values, net_label='Net Change',
                     rotation_value=45, green_color='#00b050', red_color='#ff0000')
plt.title('Waterfall Chart for First Row Data')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```

```python
# prompt: make a funnel chart having the following columns:Total Number of Family members,Members with age less than 5 year old,Members with age 5 - 17 years old,Total number of family members employed for the first row in the chart.

!pip install plotly

import plotly.graph_objects as go

# Extract data for the first row
row_data = df.iloc[0][['Total Number of Family members', 'Members with age less than 5 year old',
                       'Members with age 5 - 17 years old', 'Total number of family members employed']]

# Define labels and values
labels = row_data.index.tolist()
values = row_data.values.tolist()

# Create funnel chart
fig = go.Figure(go.Funnel(
    y=labels,
    x=values,
    textinfo="value+percent initial"))

fig.update_layout(title_text="Funnel Chart for First Row Data")
fig.show()
```

```python
# prompt: make a donut chart for Household Head Class of Worker

# Count the occurrences of each class of worker
worker_counts = df['Household Head Class of Worker'].value_counts()

# Create the donut chart
fig = go.Figure(data=[go.Pie(labels=worker_counts.index, values=worker_counts.values, hole=.4)])
fig.update_layout(title_text="Donut Chart for Household Head Class of Worker")
fig.show()
```
